```python
import pandas as pd
import numpy as np
from pyts.decomposition import SingularSpectrumAnalysis

X = pd.read_csv(
    "D:/Documents/python_projects/Netology/venv/times_series/1_introductions_with_times_series/1._Лекция_tsml/Series/weekly-closings-of-the-dowjones-.csv"
).dropna().values[:, 1]

window_size = 10
ssa = SingularSpectrumAnalysis(window_size=window_size, groups=None)
X_ssa = ssa.fit_transform(X.reshape(1, -1))

import matplotlib.pyplot as plt

plt.figure(figsize=(26, 40))
for i, component in enumerate(X_ssa[0]):
    plt.subplot(X_ssa.shape[1], 1, i+1)
    plt.plot(component)
    plt.title(f'Component {i+1}')
plt.tight_layout()
plt.show()
```
